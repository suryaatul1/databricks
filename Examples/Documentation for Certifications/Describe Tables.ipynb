## DESCRIBE EXTENDED vs DESCRIBE FORMATTED

### Official Answer (from Databricks Docs)
> "**For compatibility, FORMATTED can be specified as a synonym for EXTENDED.**"

**Bottom line:** They are **identical** - FORMATTED is just an alias for EXTENDED.

---

## Why Two Keywords Exist?

**FORMATTED** exists for **backward compatibility** with older Hive/Spark syntax:
* Legacy Hive used `DESCRIBE FORMATTED` 
* Modern Databricks/Spark standardized on `DESCRIBE EXTENDED`
* Both keywords supported so old scripts don't break

---

## What Both Commands Show

Both `DESCRIBE EXTENDED` and `DESCRIBE FORMATTED` return:

1. **Column Information**
   * Column names, data types, and comments
   
2. **Delta Statistics** (for Delta tables)
   * Statistics columns used for optimization
   
3. **Table Metadata**
   * Catalog, database, table name
   * Created time, owner
   * Table type (MANAGED vs EXTERNAL)
   * Location (storage path)
   * Provider (delta, parquet, etc.)
   
4. **Table Properties**
   * Delta features enabled
   * Compression codecs
   * Custom properties
   
5. **Storage Information**
   * Statistics (size, row count)
   * Collation
   * Predictive optimization settings

---

## Command Comparison

| Command | What You Get | Use Case |
|---------|--------------|----------|
| `DESCRIBE table` | **Columns only** | Quick schema check |
| `DESCRIBE EXTENDED table` | **Full metadata** | Complete analysis |
| `DESCRIBE FORMATTED table` | **Full metadata** (same as EXTENDED) | Compatibility with old scripts |
| `DESCRIBE EXTENDED table AS JSON` | **JSON format** | Programmatic parsing |

---

## Syntax Options

```sql
-- All of these are valid:
DESCRIBE TABLE as_catalog.as_is.books;           -- Columns only
DESCRIBE EXTENDED as_catalog.as_is.books;        -- Full metadata
DESCRIBE FORMATTED as_catalog.as_is.books;       -- Full metadata (same)
DESC EXTENDED as_catalog.as_is.books;            -- Short form

-- Get output as JSON (for scripting)
DESCRIBE EXTENDED as_catalog.as_is.books AS JSON;
DESCRIBE FORMATTED as_catalog.as_is.books AS JSON; -- Same
```

**Note:** `AS JSON` only works with EXTENDED/FORMATTED, not basic DESCRIBE.

---

## For Certification Exam

✅ **Know this:**
* FORMATTED = EXTENDED (they're synonyms)
* FORMATTED exists for backward compatibility only
* Both return identical information
* Without EXTENDED/FORMATTED, you only get columns
* JSON output requires EXTENDED or FORMATTED keyword

❌ **Don't assume:**
* That FORMATTED provides "better formatting" (it doesn't)
* That one is newer/older than the other (both supported)
* That output differs between them (it's identical)

## Alternative Commands for Table Analysis

### Show Table Properties
```sql
-- Get just the table properties
SHOW TBLPROPERTIES as_catalog.as_is.books;
```

### Delta Table History
```sql
-- See table history (Delta tables only)
DESCRIBE HISTORY as_catalog.as_is.books;
```

### System Catalog Query
```sql
-- Query Unity Catalog system tables
SELECT * FROM system.information_schema.tables 
WHERE table_catalog = 'as_catalog' 
  AND table_schema = 'as_is' 
  AND table_name = 'books';

-- Get column details
SELECT * FROM system.information_schema.columns
WHERE table_catalog = 'as_catalog' 
  AND table_schema = 'as_is' 
  AND table_name = 'books';
```

### Show Table Extended
```sql
-- Alternative command with partition details
SHOW TABLE EXTENDED IN as_catalog.as_is LIKE 'books';
```
