```python
dataset_path='/Volumes/dbacademy/labuser14322013_1774589523/bookstore'

```

```python
import dlt

@dlt.table
def order_bronze():
    return (
        spark.readStream.format("cloudFiles")
        .option("cloudFiles.format", "parquet")
        .option("cloudFiles.inferColumnTypes", "true")
        .option("cloudFiles.schemaHints", "order_id STRING, order_timestamp LONG, customer_id STRING, quantity LONG")
        .load(f"{dataset_path}/orders-raw/")
        .drop("_rescued_data")
    )
```

```python
import dlt
@dlt.table
def customers_bronze():
    return (
        spark.readStream.format("cloudFiles")
        .option("cloudFiles.format", "json")
        .load("/Volumes/dbacademy/labuser14322013_1774589523/bookstore/customers-json")
    )
```
